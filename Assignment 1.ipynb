```python
y = y_train.to_numpy()
y00 = y.reshape(len(y), 1)
X_train.shape, y.shape, y00.shape
```

```python
class LinearRegression:
    def __init__(self, X, y, learning_rate, epoch):
        self.X = X
        self.y = y
        self.learning_rate = learning_rate
        self.epoch = epoch
        self.w = None
        self.b = None
    
    def linear_regression_model(self):
        lenw = self.X.shape[1]
        self.w = np.zeros((lenw, 1))
        self.b = 0
    
        y0 = self.y.reshape(len(self.y), 1)
        
        for i in range(self.epoch):
            z = np.dot(self.X, self.w) + self.b
            dz = 1 / (len(self.X)) * (z - y0)
            dw = np.dot(self.X.T, dz)
            db = np.sum(dz)
            old_w = self.w
            old_b = self.b
    
            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db
    
            if ((np.abs(self.w - old_w) < 0.001).all() and (np.abs(self.b - old_b) < 0.001).all()) or (np.abs(dw) == 0).all() or (np.abs(db) == 0).all():
                print(f"Converged after {i + 1} iterations")
                break   
        return self.w, self.b
     
    def predict(self):
        return np.dot(self.X, self.w) + self.b

    def error(self):
        y_hat = self.predict()
        return np.mean(np.abs(self.y - y_hat))

    def r(self):
        y_hat = self.predict()
        return r2_score(self.y, y_hat)

    def plotting(self):
        for i in range(self.X.shape[1]):
            first_column = self.X[:, i]
            y_graph = [np.dot(self.w[0], x) + self.b for x in first_column]
            
            plt.scatter(first_column, self.y, marker='o', color='blue', label = 'Scatter')
            plt.plot(first_column, y_graph, marker='o', color='red', label = 'Linear function')

            plt.xlabel(str(i)+'-axis Label')
            plt.ylabel('Y-axis Label')
            plt.title('Scatter Plot')
            
            plt.legend()

            plt.grid(True)
            plt.show()
```

```python
y = y_train.to_numpy()
model = LinearRegression(X_train, y, learning_rate=0.05, epoch=2000)
model.linear_regression_model()

y_hat = model.predict()

mae = model.error()
print("Mean Absolute Error (MAE):", mae)

r_squared = model.r()
print("R-squared (R²):", r_squared)

#model.plotting()
```
